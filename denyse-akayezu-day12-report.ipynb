````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting
````

Exercise 1:

Create a simple line plot for y = x for x values ranging from 0 to 10.



````python
# get 10 linearly spaced points in the interval [0, 10]
x = np.linspace(0, 10, 20)
y = x 

# create a figure/canvas of desired size
plt.figure(figsize=(8, 6))

# plot values; with a color `red`
plt.plot(x, y, 'r')

# give labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# give a title to the plot
plt.title(r"Plot of $y=x$")
plt.show()
````

````python
### Exercise 2:
Plot `y = x^2` using Matplotlib for `x` values between -10 and 10.
````

````python
# get 10 linearly spaced points in the interval [-10, 10)
x = np.linspace(-10,10 , 50)
y = x ** 2

# create a figure/canvas of desired size
plt.figure(figsize=(8, 6))

# plot values; with a color `red`
plt.plot(x, y, 'r')

# give labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# give a title to the plot
plt.title(r"Plot of $y=x^2$")
plt.show()
````

````python
### Exercise 3:
Create a scatter plot with 50 random points where x and y values are between 0 and 10.
````

````python
plt.figure(figsize=(8, 6), dpi=60)

n = 50
X = np.random.normal(0,10,n)
Y = np.random.normal(0,10,n)
plt.scatter(X,Y, s=75, c= "blue", alpha=.5)
plt.show()
````

````python
### Exercise 4:
Plot a sine wave using Matplotlib for `x` values between 0 and 2π.
````

````python
X = np.linspace(0*np.pi, 2*np.pi, 30, endpoint=True)
S=np.sin(X)
plt.plot(X,S)
plt.show()
````
